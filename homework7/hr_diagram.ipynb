```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("3.1")

data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows = 1)

print('import of data complete')
```

```python

print("3.2")


print(f"Shape of the data: {data.shape}")

print(f"Number of rows (total stars): {data.shape[0]}\n")

magnitude = data[:, 0]    

temperature = data[:, 1] 


min_mag = np.min(magnitude)

max_mag = np.max(magnitude)


star_min_mag = np.argmin(magnitude) 

star_max_mag = np.argmax(magnitude)

print(f"MAGNITUDE:")

print(f"Minimum: {min_mag} (Corresponds to Star {star_min_mag})")

print(f"Maximum: {max_mag} (Corresponds to Star {star_max_mag})\n")


min_temp = np.min(temperature)

max_temp = np.max(temperature)

star_min_temp = np.argmin(temperature)

star_max_temp = np.argmax(temperature)

print(f"Temperature:")

print(f"Minimum: {min_temp} K (Corresponds to Star {star_min_temp})")

print(f"Maximum: {max_temp} K (Corresponds to Star {star_max_temp})\n")


print("Brightness:")

print(f"The BRIGHTEST star is Star {star_min_mag} because it has the LOWEST magnitude ({min_mag}).")

print(f"The DIMMEST star is Star {star_max_mag} because it has the HIGHEST magnitude ({max_mag}).")
```

```python
print("3.3")


dimmest_idx = np.argmax(magnitude)   

hottest_idx = np.argmax(temperature) 

coldest_idx = np.argmin(temperature) 

log_magnitude = np.log(magnitude + 6) 

log_temperature = np.log(temperature)

plt.figure(figsize=(8, 6))


scatter = plt.scatter(log_temperature, log_magnitude, c=temperature, cmap='coolwarm', label='Stars', alpha=0.7)

plt.colorbar(scatter, label='Actual Temperature (K)')


plt.gca().invert_xaxis()

plt.gca().invert_yaxis()


plt.scatter(log_temperature[brightest_idx], log_magnitude[brightest_idx], 
            color="black", marker="*", s=300, label="Brightest")


plt.scatter(log_temperature[dimmest_idx], log_magnitude[dimmest_idx], 
            color="purple", marker="*", s=300, label="Dimmest")


plt.scatter(log_temperature[hottest_idx], log_magnitude[hottest_idx], 
            color="green", marker="*", s=300, label="Hottest")


plt.scatter(log_temperature[coldest_idx], log_magnitude[coldest_idx], 
            color="orange", marker="*", s=300, label="Coldest")


plt.legend(loc='best')


plt.title("Hertzsprung-Russel Diagram", fontsize=16)

plt.xlabel("Temperature (log K)", fontsize=12)


plt.ylabel("Magnitude", fontsize=12)


plt.show()
```
