```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("2")

def make_sine_wave(x, A, w):
     # Returns the sine wave array y = A * sin(w * x)
    return A * np.sin(w * x)

x = np.linspace(0, 2 * np.pi, 1000)


amplitude = [0.5, 1, 1.5, 2, 2.5]

frequency = [1, 2, 3, 4, 5]


fig, axs = plt.subplots(1, 3, figsize=(25, 5))


for A, w in zip(amplitude, frequency):
    
  
    y_amp = make_sine_wave(x, A, 1)
   
    axs[0].plot(x, y_amp, label=f'A={A}')
    
   
    y_freq = make_sine_wave(x, 1, w)
   
    axs[1].plot(x, y_freq, label=f'ω={w}')
    
    
    y_both = make_sine_wave(x, A, w)
   
    axs[2].plot(x, y_both, label=f'A={A}, ω={w}')

titles = ['Varying Amplitude (ω=1)', 'Varying Frequency (A=1)', 'Varying Both']

for i in range(3):
  
    axs[i].set_title(titles[i], fontsize=18)
  
    axs[i].set_xlabel('x (radians)', fontsize=14)
  
    axs[i].set_ylabel('Amplitude', fontsize=14)
  
    axs[i].legend(fontsize=12, loc='upper right')
   
    axs[i].grid(True, linestyle='--', alpha=0.6) 
    


plt.tight_layout()


plt.show()
```
