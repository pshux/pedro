```python
print("4")



def is_prime(n):
     # Checks if a single number is prime.
    
    if n < 2: return False
    
    for i in range(2, int(n**0.5) + 1):
    
        if n % i == 0:
    
            return False
    
    return True

def find_prime_rows(arr):
    # Apply the is_prime check to every element in the array
    
    check_primes = np.vectorize(is_prime)
    
    prime_ = check_primes(arr)
    
    # Find rows where at least one value is True
    # axis=1 tells NumPy to look across each row
    row_has_prime = np.any(prime_, axis=1)
    
    # Use that boolean list to filter the original rows
    return arr[row_has_prime]


arr = np.array([
    [2, 3, 5],    
    [14, 6, 8],  
    [11, 13, 17], 
    [7, 10, 13]   
])

print("Rows with at least one prime:\n", find_prime_rows(arr))
```

```python
print("6")

import numpy as np

def expansion(words_array, num_spaces):
    
    # Multiplying a string by an integer repeats it. 
    # If num_spaces is 2, this becomes "  ".
    spacer = " " * num_spaces
    
   
    # The .join() method takes our spacer and inserts it between every letter of the word.
    def expand_single_word(word):
        return spacer.join(word)
        
    # Convert our single-word function into a function that works on a whole NumPy array.
    vectorized_expand = np.vectorize(expand_single_word)
    
  
    return vectorized_expand(words_array)


universe = np.array(['galaxy', 'clusters'])

print("Expansion with 1 space:")

print(expansion(universe, 1))

print("\nExpansion with 2 spaces:")

print(expansion(universe, 2))

```

```python
print("7")


import numpy as np

def secondDimmest(stars):
    
    # Sort the array along the rows (axis=1) in ascending order
    sorted_stars = np.sort(stars, axis=1)

    return sorted_stars[:, -2]


# Using a seed ensures we generate the same random numbers for testing
np.random.seed(123)

stars_array = np.random.randint(500, 2000, (5, 5))

print("Original Star Clusters:\n", stars_array)


print("\nSecond Brightest/Dimmest Stars:")

print(secondDimmest(stars_array))
```
