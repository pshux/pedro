```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import ssl
```

```python
print("2 Pandas DataFrame")

df = pd.read_csv("../../course_assignments/homework8/state_data.csv")


print("Import data complete")
```

```python
rows, columns = df.shape

print(f"The shape is {df.shape}")

print(f"4. The shape has {rows} rows and {columns} columns \n")

print("5. First 5 rows: ")

print(df.head())

print("Last 5 rows: ")

print(df.tail())

print("• The first column is the state abbreviation")
print("• The second column is the latitude, which represents represents how fat north/south the state is")
print("• The third column is the longitude which represents how far east/ west the the state is")
print("• The last column is the full name of the state")
print(f"• There are {rows} entries")
print(f"• Beyond the 50 states regularly listed, Washington DC is included (federal district) and US territories, Puerto Rico is a US territory, among others.")

med_lat = df['latitude'].median()

med_lon = df['longitude'].median()

print(f"6. Median latitude {med_lat}, Median longitude {med_lon}")

```

```python
print("7. Computing the absolute value distance to the median latitude for each state and inserting it into data frame.")

df['dist_to_med_lat'] = (df['latitude'] - med_lat).abs()
```

```python
print("8. Computing the absolute value distance to the median longitude for each state and inserting it into data frame")

df['dist_to_med_lon'] = (df['longitude'] - med_lon).abs()
```

```python
rows, columns = df.shape

print(f"9.The new shape of the data is frame{df.shape}")


print(f"The new data frame has {rows} rows and {columns} columns")
```

```python
print("10. Distance Calculations:")

closest_lat_idx = df['dist_to_med_lat'].idxmin()

farthest_lat_idx = df['dist_to_med_lat'].idxmax()

print(f"  • Closest to median latitude: {df.loc[closest_lat_idx, 'state']} by {df['dist_to_med_lat'].min():.2f} degrees")

print(f"  • Farthest from median latitude: {df.loc[farthest_lat_idx, 'state']} by {df['dist_to_med_lat'].max():.2f} degrees")


closest_lon_idx = df['dist_to_med_lon'].idxmin()

farthest_lon_idx = df['dist_to_med_lon'].idxmax()

print(f"  • Closest to median longitude: {df.loc[closest_lon_idx, 'state']} by {df['dist_to_med_lon'].min():.2f} degrees")

print(f"  • Farthest from median longitude: {df.loc[farthest_lon_idx, 'state']} by {df['dist_to_med_lon'].max():.2f} degrees\n")

print("11. Pandas documentation:")

top_3_farthest_lat = df.nlargest(3, 'dist_to_med_lat')

print("The 3 states furthest from the median latitude are:")

print(top_3_farthest_lat[['state', 'dist_to_med_lat']])
```

```python
print("3.Seaborn")

ssl._create_default_https_context = ssl._create_unverified_context

df = sns.load_dataset('mpg')

df = df.dropna(axis=1)

print("5. Dropped column with NaN values")

print(" • Dropped the column with Horsepower\n")

print(f"6. Shape of the dataset: {df.shape}")

print(f" •Rows: {df.shape[0]}, Columns: {df.shape[1]}\n")

print(" •--- First Five Rows ---")

print(df.head())

print("\n •--- Last Five Rows ---")

print(df.tail())

print(" • List the column names and describe what it represents\n")

print(" • mgg: The field efficiency of hte vehicle in miles per gallon \n")
print(" • cylinder: The number of cylinders in the carts engines\n")
print(" • displacement: The total volume of all the cylinders in the engine, which is measured in cubic inches\n")
print(" • weight: The overall weight of the car, measured in pounds\n")
print(" • acceleration: The time it takes for the car to accelerate from 0 to 60mph, measured in seconds\n")
print(" • model_year: The year the car was manufactured\n")
print(" • origin: The region where the car was manufacture\n")
print(" • name: The make and model name of the vehicle\n")

```

```python

print("7. Create custom subplots and 8. Add information\n")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Top-Left: Scatter Plot 1 (Weight vs. MPG)


sns.scatterplot(data=df, x='weight', y='mpg', hue='origin', ax=axes[0, 0])

axes[0, 0].set_title("Fuel Efficiency by Vehicle Weight", fontsize=14)

axes[0, 0].set_xlabel("Weight (lbs)", fontsize=12)

axes[0, 0].set_ylabel("Fuel Efficiency (mpg)", fontsize=12)

# Top-Right: Scatter Plot 2 (Displacement vs. Acceleration)

sns.scatterplot(data=df, x='displacement', y='acceleration', hue='cylinders', palette='viridis', ax=axes[0, 1])

axes[0, 1].set_title("Acceleration vs. Engine Displacement", fontsize=14)

axes[0, 1].set_xlabel("Engine Displacement (cu. inches)", fontsize=12)

axes[0, 1].set_ylabel("Acceleration (0-60 mph in seconds)", fontsize=12)

# Bottom-Left: Histogram (MPG Distribution)

sns.histplot(data=df, x='mpg', bins=20, color='purple', kde=True, ax=axes[1, 0])

axes[1, 0].set_title("Distribution of Fuel Efficiency", fontsize=14)

axes[1, 0].set_xlabel("Fuel Efficiency (mpg)", fontsize=12)

axes[1, 0].set_ylabel("Frequency (Number of Cars)", fontsize=12)

# Bottom-Right: Categorical Boxplot (Origin vs. MPG)

sns.boxplot(data=df, x='origin', y='mpg', hue='origin', palette='Set2', legend=False, ax=axes[1, 1])

axes[1, 1].set_title("Fuel Efficiency Spread by Region", fontsize=14)

axes[1, 1].set_xlabel("Region of Origin", fontsize=12)

axes[1, 1].set_ylabel("Fuel Efficiency (mpg)", fontsize=12)

plt.tight_layout()

plt.show()

print("9. Explain what data each of the subplots represent and why these columns ? What was learned about the data from each of the subplots ?")

print(" • Top-Left subplot: This plots the weight of the car on the x-axis against its miles per gallon on the y-axis.\n")

print("    - What it represents/Why: This was plotted because weight is the limiting factor for fuel efficiency.\n ")

print("    - What I learned: There is a negative correlation between weight and mpg. As cars get heavier, their fuel efficiency drops.\n ")

print(" • Top-Right subplot: This plots the engine size  of the car on the x-axis against acceleration on the y-axis.\n")

print("    - What it represents/Why: This was plotted in order to compare if larger engine gives faster acceleration.\n ")

print("    - What I learned: Larger engines does not have drastically faster 0-60 times in the data set. The acceleration times are clustered between 10 and 20 seconds.\n ")

print(" • Bottom-Left histogram: This plots Fuel efficiency on the x-axis against the numbers of cars on the y-axis.\n")

print("    - What it represents/Why: This shows the distribution of fuel efficiency across every single car in the dataset. This was chose to under stand the efficiency for the the cats in the 1970s and early 1980s.\n ")

print("    - What I learned: The data is skewed to the right The majority of the cars from this timeline had an efficiency between 13 and 20 mpg.\n ")

print(" • Bottom-Right Box: This plots the region or origin of manufacture against its fuel efficiency .\n")

print("    - What it represents/Why: This categorical plot shows the spread of fuel efficiency grouped by where the car was made. This was plotted to show which region produced the most economical cars .\n ")

print("    - What I learned: Japan  produces the most fuel efficient vehicles in this timeline with the highest median mpg. The United States has the lowest median. A few clusters of gas-guzzling vehicles and few outliers.\n ")

```
