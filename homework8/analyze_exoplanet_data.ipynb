```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
print("4.2")

df = pd.read_csv('exoplanet_data.csv', comment = '#')

print(f"\nThe shape of the data set {df.shape}")

columns_names = df.columns.tolist()

print("\nColumn Names:")
print(f"\n{columns_names}")

print(f"\nFive columns do not know what they mean, sy_disterr1, st_logg, st_logger1, st_loggerr2,sv_vmag ")


print("\nTable Label\Database Column Name: description of parameter; think the parameter means")

print("\nPlanet name\pl_name: Planet name most commonly used in literature; planet that was named when discovered ")

print("\nHost Name\hostname: Stellar name most commonly used in the literature; the host star, or star that planets orbit")

print("\nDiscovery Method\discoverymethod: Method by which the planet was first identified; the method used to find planet")

print("\nSpectral Type\st_spectype: Classification of the star based on their spectral characteristics following the Morgan-Keenan system; the measurement of the temperature of a star")

print("\nEquilibrium Temperature [K]\pl_eqt: The equilibrium temperature of the planet as modeled by a black body heated only by its host star, or for directly imaged planets, the effective temperature of the planet required to match the measured luminosity if the planet were a black body\the temperature it would need to have to produce the observed brightness if it were a blackbody")

print("\nPlanet Mass or Mass*sin(i) [Jupiter Mass]\pl_msinij: Minimum mass of a planet as measured by radial velocity, measured in units of masses of Jupiter\minimum possible mas of a planet")

print("\nNote: for Planet Mass or Mass*sin(i) [Jupiter Mass], in the website the column name is pl_msinij, but in .csv file, it is listed as pl_bmassj")


columns_to_keep = ['pl_name', 'hostname', 'discoverymethod', 'st_spectype', 'pl_eqt', 'pl_bmassj']

reduced_df = df[columns_to_keep]


reduced_df = reduced_df.dropna()


print("\nShape of the 'smaller' dataset:")
print(reduced_df.shape)



print("\nFirst 5 rows:")
print(reduced_df.head())

print("\nLast 5 rows:")
print(reduced_df.tail())


print("\nColumn Titles:")
print(reduced_df.columns.tolist())


```

```python
print("4.4")


discovery_counts = reduced_df['discoverymethod'].value_counts()

print(discovery_counts)

print("\nThe discovery methods are Transit, Radial velocity, and Imaging")

print("\nThere 55 for Transit, 13 for Radial Velocity and 2 for Imaging")

print("\n The one with the most is Transit and the one with the least is Imaging")

print("\n Transit Method: This method involves looking if a planets orbit passes between the star and the Earth, it will block a fraction of the stars light")

print("\n Radial Velocity: Observing the changes in wavelengths for certain spectrum of light due to gravity")

print("\n Imaging: Basically taking a picture, but difficult because host star is brighter than the planet")





plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=reduced_df, 
    x='pl_bmassj',           
    y='pl_eqt',              
    hue='discoverymethod'    
)

plt.title('Planet Mass vs. Equilibrium Temperature')
plt.xlabel('Planet Mass (Jupiter Masses)')
plt.ylabel('Equilibrium Temperature (K)')
plt.legend(title='Discovery Method', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()



plt.figure(figsize=(8, 6))


sns.scatterplot(
    data=reduced_df, 
    x=np.log(reduced_df['pl_bmassj']),   
    y=np.log(reduced_df['pl_eqt']),      
    hue='discoverymethod'             
)


plt.title('Log of Planet Mass vs. Log of Equilibrium Temperature')
plt.xlabel('Planet Mass')
plt.ylabel('Equilibrium Temperature')

print("\nLog-Log plot\n")
plt.legend(title='Discovery Method', bbox_to_anchor=(1.05, 1), loc='upper left')


plt.show()

print("According to the graph(s), the exoplanets found by the imaging method are more likely to be bigger in mass.")



massive_planets = reduced_df[reduced_df['pl_bmassj'] > 5]

coldest_massive_planet = massive_planets.sort_values(by='pl_eqt').head(1)


print("\nPrinted entire row of coldest planet with mass larger than five times Jupiter Masses and answer to questions\n")

print('\n', coldest_massive_planet)



name = coldest_massive_planet['pl_name'].values[0]

index = coldest_massive_planet.index[0]

temp = coldest_massive_planet['pl_eqt'].values[0]

mass = coldest_massive_planet['pl_bmassj'].values[0]

star_type = coldest_massive_planet['st_spectype'].values[0]

print(f"\nName: {name}")
print(f"Index: {index}")
print(f"Equilibrium Temperature: {temp} K")
print(f"Mass: {mass} Jupiter masses")
print(f"Star Type: {star_type}")
print("Is it a sun-like star? No, K-type stars are cooler than G-type Sun.")


print("\n Update lo-Log plot pointing out the coldest_massive_planet\n")
plt.figure(figsize=(8, 6))


sns.scatterplot(
    data=reduced_df, 
    x=np.log(reduced_df['pl_bmassj']),   
    y=np.log(reduced_df['pl_eqt']),      
    hue='discoverymethod',
    alpha=0.5 
)


plt.scatter(
    x=np.log(coldest_massive_planet['pl_bmassj']), 
    y=np.log(coldest_massive_planet['pl_eqt']),    
    color='red',                                   
    marker='*',                                   
    s=400,                                        
    edgecolor='black',                             
    label=coldest_massive_planet['pl_name'].values[0] 
)

plt.title('Log of Planet Mass vs. Log of Equilibrium Temperature')
plt.xlabel('Planet Mass')
plt.ylabel('Equilibrium Temperature')


plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')


plt.show()


print("Summarize abstract: Astronomers used the James Webb Space Telescope (JWST) to directly image a giant exoplanet orbiting the star Epsilon Indi A. While earlier indirect measurements (observing the star 'wobble') hinted at a planet's existence, the actual planet they photographed has very different properties than previously predicted. It is cold (around 275 K) and faint at certain infrared wavelengths. This faintness suggests its atmosphere might be rich in heavy elements like carbon or completely covered in a thick layer of clouds.")

print("\nDoes you answer math the exoplanet: Yes! The abstract explicitly states they refer to this newly imaged planet as eps Ind A b, which is the exact same planet isolated from the dataset.")


print("\nWhat is special about the planet you found: It is one of the coldest exoplanets ever directly photographed (sitting right around 2°C, near the freezing point of water). Usually, direct imaging only works on very young, extremely hot planets that glow brightly. Being able to directly see a mature, freezing-cold gas giant orbiting a star with an age similar to the Sun is a massive milestone for astronomy. Additionally, the planet defied astronomers' expectations regarding its orbit and mass")

print("\nDo the mass and equilibrium temperature math what the paper reported ? Yes, they are a match! Temperature: The abstract explicitly reports a temperature of approximately 275 K, which matches the 275.0 found in the pl_eqt column. Mass: The paper describes the new JWST data as being consistent with a 'super-Jupiter' having a mass roughly six times the mass of Jupiter. This perfectly aligns with the specific value of 6.31 Jupiter masses (pl_bmassj) that was pulled from the dataset,")


```
